```python
def estimate_noise(I):

  H, W = I.shape

  M = [[1, -2, 1],
       [-2, 4, -2],
       [1, -2, 1]]

  sigma = np.sum(np.sum(np.absolute(convolve2d(I, M))))
  sigma = sigma * math.sqrt(0.5 * math.pi) / (6 * (W-2) * (H-2))

  return sigma
```

```python
noise_init_total = array.array('f')
noise_fin_total = array.array('f')
obj_count = 0
for i in range(1,14):

  #i=+1
  path=file+str(i)+'.mp4'
  print('Video File Data ID' + path)
  capture = cv2.VideoCapture(path)
  Video_FILE = path

  vidcap=cv2.VideoCapture(Video_FILE)
  #print('Initial Number of Frames: ' + str(int(capture.get(cv2.CAP_PROP_FRAME_COUNT))))

  success,image = vidcap.read()
  count = 0
  while success and count<20:
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    noise_init=estimate_noise(gray_image)
    noise_init_total.append(noise_init)
    #print('Initial Noise: ' + str(noise_init))
    #print('Before: Number of Objects: ' + str(Obj_Detect(image)))


    dat = cv2.fastNlMeansDenoisingColored(image,None,10,10,7,21)
    gray_image = cv2.cvtColor(dat, cv2.COLOR_BGR2GRAY)
    noise_fin=estimate_noise(gray_image)

    noise_fin_total.append(noise_fin)


    #print('Reduced Noise: ' + str(noise_fin))
    #print('After: Number of Objects: ' + str(Obj_Detect(image)))



    count+=1
  print('Mean Inital Noise: ' + str(mean(noise_init_total)))
  print('Mean Reduced Noise: ' + str(mean(noise_fin_total)))
```
