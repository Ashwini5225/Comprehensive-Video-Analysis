```python
for i in range(1,5):


  path=file+str(i)+'.mp4'
  print('Video File Data ID' + path)
  capture = cv2.VideoCapture(path)
  Video_FILE = path

  vidcap=cv2.VideoCapture(Video_FILE)
  print('Initial Number of Frames: ' + str(int(capture.get(cv2.CAP_PROP_FRAME_COUNT))))

  #p_frame_thresh = 90000
  p_frame_thresh = random.randrange(90000,99999)
  total_number_obj=0

  ret, prev_frame = vidcap.read()

  kf_count=0

  while ret:
    ret, curr_frame = vidcap.read()
    #print('Number of Objects: ' + str(Obj_Detect(curr_frame)))


    if ret:
        diff = cv2.absdiff(curr_frame, prev_frame)
        non_zero_count = np.count_nonzero(diff)
        if non_zero_count > p_frame_thresh:
            #print('Got P-Frame')
            kf_count=kf_count+1
        prev_frame = curr_frame
        #total_number_obj+=Obj_Detect(curr_frame) / kf_count

  print('Final Number of Key Frames: '+str(kf_count))
  print('Final Adaptive Threshold: '+str(p_frame_thresh))
```
